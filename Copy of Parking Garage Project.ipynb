# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP).

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
# Your Code will go here
class ParkingGarage:
    def __init__(self, currentTicket=None, tickets=None, parkingSpaces=None):
        self.tickets = tickets if tickets is not None else [1, 2, 3, 4, 5]
        self.parkingSpaces = parkingSpaces if parkingSpaces is not None else [1, 2, 3, 4, 5]
        self.currentTicket = currentTicket if currentTicket is not None else {}

    def takeTicket(self):
        if len(self.tickets) > 0 and len(self.parkingSpaces) > 0:
            self.tickets.pop()
            self.parkingSpaces.pop()
            self.currentTicket["Paid"] = False
            print("Take ticket and proceed. ")
        else:
            print("Sorry, this parking lot is full")


    def payForParking(self):
        try:
            amount = float(input("Enter your payment please: "))
            if amount >= 5:
                print("Ticket has been paid. You have 15 minutes to leave.")
                self.currentTicket["Paid"] = True
            else:
                print("Insufficient amount. Please check your ticket and update your payment.")
        except ValueError:
            print("Invalid input. Please enter a valid amount.")

    def leaveGarage(self):
        if self.currentTicket.get("Paid", False):
            print("Thank you, have a great day!")
            self.tickets.append(1)
            self.parkingSpaces.append(1)
            self.currentTicket["Paid"] = False
        else:
            print("Please enter correct payment.")

def run():
    myLot = ParkingGarage()
    while True:
        response = input("Welcome to our parking garage. What would you like to do? Enter: Enter, Pay, Exit: ").strip().lower()

        if response == 'exit':
            myLot.leaveGarage()
            break
        elif response == 'pay':
            myLot.payForParking()
        elif response == 'enter':
            myLot.takeTicket()
        else:
            print("Invalid response. Please enter 'Enter', 'Pay', or 'Exit'.")

run()

  #create a run statement to only execute if the amount of spaces and tickets
  #availble is less than the amount of spaces and tickets available
```
