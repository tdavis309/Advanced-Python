```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019 - AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

n_group = air_bnb.groupby(['neighbourhood_group'], as_index = False).count()
n_group

#there are 5 neighborhood groups, Manhattan shows up more
```

```python
air_bnb.groupby('neighbourhood_group').describe()
```

```python
air_host= air_bnb.groupby(["host_id", "host_name"])
bussiest_host = air_host['number_of_reviews'].sum().sort_values(ascending = False).head(5)
bussiest_host

```

```python
#Which neighorhood group has the highest average price?
air_neighborhood= air_bnb.groupby("neighbourhood_group")
ave_neighborhood_price = air_neighborhood["price"].mean().sort_values(ascending = False).round()
ave_neighborhood_price
```

```python
# Which neighbor hood group has the highest total price?
max_neighborhood_price = air_neighborhood["price"].sum().sort_values(ascending = False).round()
max_neighborhood_price
```

```python
#Which top 5 hosts have the highest total price?
highest_host = air_bnb.groupby(["host_id", "host_name"])
highest_host_price = highest_host['price'].sum().sort_values(ascending = False).round()
highest_host_price.head(5)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

zero_availability = air_bnb[(air_bnb['availability_365'] == 0) & (air_bnb['number_of_reviews'] >= 100)].reset_index(drop = True)
zero_availability_host = zero_availability[["host_id", "host_name", "availability_365"]]
zero_availability_host
```

```python
# What host has the highest total of prices and where are they located?
highest_host1 = air_bnb.loc[air_bnb["host_id"] == 219517861]
highest_host1[["host_name", "neighbourhood_group", "neighbourhood" ]]
#Sonder, from Manhattan
```

```python
# When did Danielle from Queens last receive a review?

danielle_queens = air_bnb[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')]
danielle_queens ["last_review"].sort_values(ascending = False)

#07-08-2019 was the last review
```

## Further Questions

1. Which host has the most listings?

```python
hosts_listings_count = air_bnb.groupby('host_id')['id'].nunique()

host_with_most_listings = hosts_listings_count.idxmax()
most_listings_count = hosts_listings_count.max()
host_with_most_listings
#host_id 219517861
```

2. How many listings have completely open availability?

```python
open_availability_listings = (air_bnb['availability_365'] == 365).sum()
open_availability_listings
```

3. What room_types have the highest review numbers?

```python
room_groups = air_bnb.groupby('room_type')
room_reviews =room_groups['number_of_reviews'].sum().sort_values(ascending=False)
highest_review_room_type = room_reviews.idxmax()
highest_review_room_type 
```
