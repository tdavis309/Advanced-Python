# What player scored the most points?

```python
nba
pts = nba.groupby("Name")["Points"].sum()
pts.sort_values(ascending=False)
#James Harden scored the most points
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
blks = nba.groupby("Name")["Blocks"].sum()
blks.sort_values(ascending=False)
#Myles turner had the most blocks
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
winshares = nba.groupby("Team")["WS"].sum()
winshares.sort_values(ascending=False)
#Milwaukee Bucks
```

# What player had the best 3-pt percentage? 

```python
nba.groupby("Name")["FG3%"].max().sort_values(ascending = False)


#the players with the best 3pt % are Eric Moreland, Scott Machando, Tevon Duval, Jordan Sibert.
```
